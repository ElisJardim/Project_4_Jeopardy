```python
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
JEOP_df = pd.read_csv('./Resources/QuestionsSQL.csv')
JEOP_df.head()
```

```python
data_Cat = JEOP_df['category'].value_counts().to_dict()
wc_Cat = WordCloud().generate_from_frequencies(data_Cat)
wc_Cat.to_file("./Images/Category_Cloud_RE.png")

plt.imshow(wc_Cat)
plt.axis('off')
plt.figure( figsize=(40,20) )
plt.show()
```

```python
data_Ans = JEOP_df['answer'].value_counts().to_dict()
wc_Ans = WordCloud().generate_from_frequencies(data_Ans)
wc_Ans.to_file("Images/Answers_Cloud_RE.png")

plt.imshow(wc_Ans)
plt.axis('off')
plt.figure( figsize=(40,20) )
plt.show()
```
